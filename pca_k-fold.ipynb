```python
# Loading datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, LabelEncoder

```

```python
# Loading the dataset
data = pd.read_csv(r'C:\Users\NGARE\DS_Python_Lux\Ds_projects\human-personality-ml-model\personality_datasert.csv')
data.head()
```

```python
# Splitting column
x = data.drop(columns="Personality", axis=1)
y = data["Personality"]
```

```python
# Label encoding
encoder = LabelEncoder()

x["Stage_fear"] = encoder.fit_transform(x["Stage_fear"])
x["Drained_after_socializing"] = encoder.fit_transform(x["Drained_after_socializing"])

x.head()
```

```python
# Scaling the dataset
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# PCA
pca = PCA(n_components=7)

pca_results = pca.fit_transform(x_scaled)

pca_results
```

```python
# creating pca graph with all the 7 columns
x_pca1 = pca.fit(x_scaled)
plt.plot(range(1, len(x_pca1.explained_variance_ratio_)+1),
         x_pca1.explained_variance_ratio_.cumsum(), marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance vs. Number of Components")
plt.grid(True)
plt.show()
```
